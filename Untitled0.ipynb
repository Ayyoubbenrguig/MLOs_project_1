```python
# preparation pour la matrice de confusion

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

**Linear regression**

```python
x_train, x_test, y_train, y_test = train_test_split(features, Y, train_size=0.8, random_state=3)
LinearReg = LinearRegression()
LinearReg.fit(x_train, y_train)
```

```python
predictions = LinearReg.predict(x_test)

LinearRegression_MAE = mean_absolute_error(y_test, predictions)
LinearRegression_MSE = mean_squared_error(y_test, predictions)
LinearRegression_R2 = r2_score(y_test, predictions)

print("Mean Absolute Error is: ", LinearRegression_MAE )
print("Mean squared Error is: ", LinearRegression_MSE )
print("R-squared  is: ", LinearRegression_R2  )
```

```python
Evaluated ={
    'Metric': ['MAE', 'MSE', 'R-squared'],
    'Value': [LinearRegression_MAE, LinearRegression_MSE, LinearRegression_R2]
}
metrics_data = pd.DataFrame(Evaluated)
```

***KNN***

```python
# We can calculate the accuracy of KNN for different values of k

Ks = 100
mean_acc = np.zeros((Ks-1))
mean_jacc = np.zeros((Ks-1))
mean_f1 = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    neigh = KNeighborsClassifier(n_neighbors = n).fit(x_train,y_train)
    yhat=neigh.predict(x_test)
    mean_acc[n-1] = accuracy_score(y_test, yhat)
    mean_jacc[n-1] = jaccard_score(y_test, yhat)
    mean_f1[n-1] = f1_score(y_test, yhat)
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
```

```python
# Plot the model accuracy for a different number of neighbors.

plt.plot(range(1,Ks),mean_acc,'green')
plt.plot(range(1,Ks),mean_jacc,'red')
plt.plot(range(1,Ks),mean_f1,'blue')

#plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
#plt.fill_between(range(1,Ks),mean_acc - 3 * std_acc,mean_acc + 3 * std_acc, alpha=0.10,color="green")
plt.legend(('Accuracy_Score', 'JaccardIndex','F1_Score'))
plt.ylabel('metric')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1)
print( "The best Jaccord-index was with", mean_jacc.max(), "with k=", mean_jacc.argmax()+1)
print( "The best F1-score was with", mean_f1.max(), "with k=", mean_f1.argmax()+1)
```

```python
k=mean_acc.argmax()+1
KNN = KNeighborsClassifier(n_neighbors=k)
KNN.fit(x_train, y_train)
```

```python
predict_KNN = KNN.predict(x_test)
KNN_Accuracy_Score = accuracy_score(y_test, predict_KNN)
KNN_JaccardIndex = jaccard_score(y_test, predict_KNN)
KNN_F1_Score = f1_score(y_test, predict_KNN)
```

```python
Evaluated_KNN ={
    'Metric': ['accuracy_score', 'jaccord_score', 'f1_score'],
    'Value': [KNN_Accuracy_Score, KNN_JaccardIndex,KNN_F1_Score ]
}
metrics_data_KNN = pd.DataFrame(Evaluated_KNN)
metrics_data_KNN
```

```python
cnf_matrix = confusion_matrix(y_test, predict_KNN, labels=[0,1])
np.set_printoptions(precision=2)

print (classification_report(y_test, predict_KNN))

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['Not-RainTomorrow','RainTomorrow'],normalize= False,  title='Confusion matrix')
```

```python
# We can calculate the accuracy of KNN for different values of random_state

nb_random =20
mean_acc_LR =[]
mean_jacc_LR=[]
mean_f1_LR = []
for i in range(1,nb_random):
  x_train, x_test, y_train, y_test = train_test_split(features, Y, train_size=0.8, random_state=i)
  LR = LogisticRegression()
  LR.fit(x_train, y_train)
  predict_LR =LR.predict(x_test)
  mean_acc_LR.append(accuracy_score(y_test, predict_LR))
  mean_jacc_LR.append(jaccard_score(y_test, predict_LR))
  mean_f1_LR.append(f1_score(y_test, predict_LR))
  LR_Log_Loss = log_loss(y_test, predict_LR)
```

```python
# Plot the model accuracy for a different number random_state.

plt.plot(range(1,nb_random),mean_acc_LR,'green')
plt.plot(range(1,nb_random),mean_jacc_LR,'red')
plt.plot(range(1,nb_random),mean_f1_LR,'blue')


plt.legend(('Accuracy_Score', 'JaccardIndex','F1_Score'))
plt.ylabel('metric')
plt.xlabel('Number of random_state')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", max(mean_acc_LR), "with random_state=", mean_acc_LR.index(max(mean_acc_LR))+1)
print( "The best Jaccord-index was with", max(mean_jacc_LR), "with random_state=", mean_jacc_LR.index(max(mean_jacc_LR))+1)
print( "The best F1-score was with", max(mean_f1_LR), "with random_state", mean_f1_LR.index(max(mean_f1_LR))+1)
```
